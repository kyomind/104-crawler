### 再開啟，轉成CSV格式
- 鍵變成第一行，也就是欄位名稱
- 其餘都是資料列

```python
with open('data/Flask-20.json','r',encoding='utf8') as file:
    jobs=json.load(file)
```

```python
jobs
```

```python
with open('data/Flask.csv','w',encoding='utf8',newline='') as file:
    csvout=csv.writer(file)
    csvout.writerow(list(jobs[0].keys()))
    for job in jobs:
        csvout.writerow(job.values())
# 就這麼簡單
# 只是不清楚為什麼都照順序就是了
```

```python
jobs[0].keys()
```

### 改為，直接使用DictWriter()寫入
做法上更直觀，應採用

```python
with open('data/FlaskDictWrite.csv','w',encoding='utf8',newline='') as file:
    csvout=csv.DictWriter(file,jobs[0])#job[0]就是一個字典
# 注意和writer的差別是這裡有2個引數
# 第二個引數吃一個list形成第一列，也就是欄位列
    
    csvout.writeheader() #把吃的第一列資料寫入
    
# 這裡最精彩，可以直接吃一個list裡面的全部字典元素，的每一個值！！！
# 不用再一個一個迭代了
# 此時要處理兩層(字典層+鍵值層，最外層list不計)，必須使用writerowS，不然就是錯誤
    csvout.writerows(jobs)
# 結果和前面相同，但寫法顯然更精簡
```

```python
jobs
```

### 再把CSV檔讀出來，並還原為字典格式！

```python
with open('data/FlaskDictWrite.csv',encoding='utf8') as file:
    csvin=csv.DictReader(file)
    jobs_fromcsv=[job for job in csvin]
```

```python
jobs_fromcsv
# 雖然和書上結果有出入，但查文件確實就是OrderedDict
```

```python
jobs_fromcsv[0]['職位名稱']
# OrderiDict還是要用鍵來存取值，依舊不是index，不要以為它是seq序列
```
