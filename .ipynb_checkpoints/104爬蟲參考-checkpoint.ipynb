```python
import requests
from bs4 import BeautifulSoup as bs
search_url = "https://www.104.com.tw/area/freshman/search?keyword=%E6%A9%9F%E5%99%A8%E4%BA%BA&area=6001001000,6001002000&jobcategory=2007000000&industry=&page=1&sortField=APPEAR_DATE&sortMode=DESC"
head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.90 Safari/537.36',
        'Accept-Language':'zh-TW,zh;q=0.8,en-US;q=0.6,en;q=0.4'} 
res = requests.get(search_url, headers = head)
print(res.encoding)
```

### response物件編碼不是UTF8，作者自己發現的，必須要改成UTF-8
- 做法也很簡單，改response物件的encoding屬性值就可以了

結果新版網頁已經是UTF-8了

```python
res.encoding = 'utf8'
soup = bs(res.text)
soup
```

### soup.select()回傳值
- 是list，跟find_all的ResultSet又有一點不同
- select是另一種尋找標籤物件的方法，原理等同於css的選擇器
- 所以我也一樣可以用find方法做，請見下述

### 注意select的回傳值是list
而因為本頁只有一個class名稱是job_box，所以list元素只有一個，元素為該div Tag物件

```python
job_box = soup.select(".job_box")
job_box
```

```python
type(job_box)
```

```python
len(job_box)
```

### 使用find()
- 第一個參數是HTML標籤，第二個參數是預設為class name
- find()回傳Tag物件，find_all()回傳ResultSet

```python
job_box_find=soup.find('div','job_box')
```

```python
job_box_find
```

```python
type(job_box_find)
```

```python
job_box_findall=soup.find_all('div','job_box')
type(job_box_findall)
```

### 全網頁只有一個class名稱叫job_box，所以用soup.select就好了
