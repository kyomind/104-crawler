```python
import requests
from bs4 import BeautifulSoup as bs
search_url = \
'https://www.104.com.tw/jobs/search/?ro=0&isnew=3&keyword=php&area=6001002000%2C6001001000&order=1&asc=0&page=1&mode=s&jobsource=2018indexpoc'
head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.90 Safari/537.36',
        'Accept-Language':'zh-TW,zh;q=0.8,en-US;q=0.6,en;q=0.4'} 
res = requests.get(search_url, headers = head)
print(res.encoding)
```

### response物件編碼不是UTF8，作者自己發現的，必須要改成UTF-8

```python
res.encoding = 'utf8'
soup = bs(res.text)
soup
```
